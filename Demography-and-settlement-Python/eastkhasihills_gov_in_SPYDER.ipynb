# east khasi hill website spyder

```python
from bs4 import BeautifulSoup
import numpy as np
import ssl
import re
#import wget
from urllib.request import Request, urlopen
# Ignore SSL certificate errors
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE
```

```python
url_main = r'https://eastkhasihills.gov.in/'
def iterator(url):
    req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
    print("attempting to spyder " + url)
    html = urlopen(req).read()
    
    soup = BeautifulSoup(html, "html.parser")

    s=[]
    # Retrieve all of the anchor tags
    tags = soup('a')

    for tag in tags:
        link = tag.get('href', None)
        if (link.startswith("http")):
            s.append(link) 
    s = np.unique(s)
    return s
```

### Iterate onto URL , get links

```python
finlist = iterator(url_main)
urllist = []
for link in finlist:
    if not link.endswith('/'):
        link += '/'
        urllist.append(link)
    else:
        urllist.append(link)
urllist = np.unique(urllist)        
print(urllist)
print(len(urllist))
```

### Iterate onto links in finlist and get dictionary of links with keys as items from finlist and values as list of webpages within key link

```python
d = dict()
for link in finlist[3:-3]:
    print("doing "+ link)
    try:
        d[link] = iterator(link)
    except:
        continue
```
